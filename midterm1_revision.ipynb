## basic functions
- write a function to add two numbers and add some basic comments
- run it once on two integers
- now run it once on an integer and a string
- can you handle this error better and return nan if one input is a string? Use a general Exception


```python
def add_numbers(foo, bar):
    """
    a function to add two numbers
    input: foo, integer; bar, integer
    output: integer
    """
    try:
        res = foo + bar
    except Exception as e:
        print(f'exception: {e}')
        #res = np.nan
    return res
```

```python
'foo' * 3
```

```python
res = add_numbers(1,'abc')
```

```python
res = add_numbers(1,2)
```

# loops
- create a dictionary with 5 students names and their fav colors
- loop over the dictionary and print keys and values
- now choose any two students and only print names and colors for them
- now choose any one student and only print their fav color
- now choose any one student and do not print their fav color. Only print for everyone else
- can you write a while loop that loops over the keys and prints them?

```python
python_class = {
    'alejandro': 'purple',
    'himat': 'green',
    'alexander': 'blue',
    'carmen': 'yellow',
    'elizabeth': 'black'
}
```

```python
python_class
```

```python
class_keys = list(python_class.keys())
```

```python
class_keys
```

```python
idx = 0
while idx < len(class_keys):
    print(class_keys[idx], python_class[class_keys[idx]])
    idx = idx + 1
```

```python
type(python_class.items())
```

```python
python_class.items()
```

```python
for k,v in python_class.items():
    print(k,v)
```

```python
for k,v in python_class.items():
    if k == 'alejandro' or k == 'himat':
        print(k,v)
    else:
        print('not the chosen student')
```

```python
for k,v in python_class.items():
    if k == 'alejandro':
        print(k,v)
    elif k == 'himat':
        print(k,v)
    else:
        print('not the chosen student')
```

```python
for k,v in python_class.items():
    if k == 'alejandro':
        print(v)
```

```python
for k,v in python_class.items():
    if k == 'alejandro':
        print(v)
        break
```

```python
for k,v in python_class.items():
    if k == 'alejandro':
        print('')
    else:
        print(v)
```

```python
for k,v in python_class.items():
    if k != 'alejandro':
        print(v)
```

```python
for k,v in python_class.items():
    if k == 'alejandro':
        continue
    print(v)
```
